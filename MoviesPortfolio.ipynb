```python
# Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 

# Read in the data
df = pd.read_csv(r'C:\Users\kwagg\OneDrive\Documents\PortfolioProject\movies.csv')
```

```python
# Looking at the data
df.head()
```

```python
df.info()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
# Check for missing data again

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Check data types for columns

df.dtypes
```

```python
# Change data type for 'Budget' and 'Gross' to integer to remove decimals

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Extract the year from 'Released' column to create new column with correct year

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head(20)
```

```python
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
df.info()
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Number of values in each column stays the same, so dataset had no duplicates

df.info()
```

```python
df.head()
```

```python
# Investigating Correlations

# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Revenue')
plt.show()
```

```python
# Plot Budget vs Gross Revenue using Seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.xlabel("Budget for Film")
plt.ylabel("Gross Revenue")
plt.title("Budget vs Gross Revenue")
```

```python
# Investigating Correlations

df.corr(numeric_only=True)
```

```python
df.corr(numeric_only=True, method='spearman')
```

```python
corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Characteristics')
plt.xlabel('Movie Characteristics')
plt.ylabel('Movie Characteristics')

plt.show()
```

```python
# Plot Score of Film vs Gross Revenue using Seaborn

sns.regplot(x='score', y='gross', data = df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.xlabel("Score of Film")
plt.ylabel("Gross Revenue")
plt.title("Score of Film vs Gross Revenue")
```

```python
# Plot Runtime of Film vs Budget using Seaborn

sns.regplot(x='runtime', y='budget', data = df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.xlabel("Runtime of Film")
plt.ylabel("Budget")
plt.title("Runtime of Film vs Budget")
```

```python
# Plot Votes of Film vs Gross Revenue using Seaborn

sns.regplot(x='votes', y='gross', data = df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.xlabel("Votes for Film")
plt.ylabel("Gross Revenue")
plt.title("Votes of Film vs Gross Revenue")
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_pairs = corr_matrix.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
